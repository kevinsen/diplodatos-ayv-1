¿Hay algún NaN en una columna que no sea NaN en la otra?

```python
pd.value_counts((np.isnan(data_id.loc[:,'pf_identity'])^np.isnan(data_id.loc[:,'hf_score'])))
```

No hay ningún valor faltante cruzado (nan en pf y ~nan en hf o viceversa). Son comparables desde ese punto de vista.
En cuanto a la comparación en sí misma, la información es acerca de si los derechos personales referidos a la propia identidad están garantizados en mayor o menor medida que el resto de los indicadores que componen hf_score.
Los valores faltantes pueden:
*Hacerse cero, pero alteran las medidas de agregación porque contribuyen a n
*Eliminarse sus filas
Los outliers se identifican como los valores que están a más de 1.5*IQ de los cuantiles 25 y 75%, con IQ=rango intercuartil

Observación de cuartiles y cálculo de rango intercuartil para encontrar outliers

```python
data_id.loc[:,['pf_identity','hf_score']].quantile(0.25)
```

```python
data_id.loc[:,['pf_identity','hf_score']].quantile(0.50)
```

```python
data_id.loc[:,['pf_identity','hf_score']].quantile(0.75)
```

```python
(data_id.loc[:,['pf_identity','hf_score']].quantile(0.75))-(data_id.loc[:,['pf_identity','hf_score']].quantile(0.25))
```

```python
IQdist = (data_id.loc[:,['pf_identity','hf_score']].quantile(0.75))-(data_id.loc[:,['pf_identity','hf_score']].quantile(0.25))
medianas = data_id.loc[:,['pf_identity','hf_score']].quantile(0.5)
print(IQdist, medianas)
```

```python
outliers_minimos = medianas - IQdist
outliers_minimos
```

```python
outliers_maximos = medianas + IQdist
outliers_maximos
```

Extracción de outliers

```python
data_id_outliers = data_id[(data_id['pf_identity']<outliers_minimos[0])|(data_id['pf_identity']>outliers_maximos[0])
                          |(data_id['hf_score']<outliers_minimos[1])|(data_id['hf_score']>outliers_maximos[1])]
```

```python
data_id_outliers.shape
```
