### 1.2- Obtener media, mediana y desviación estándar de las variables *pf_identity* y *hf_score* en el mundo y compararla con la de Latinoamérica y el caribe. ¿Tiene sentido calcular la moda?

```python
pd.concat([ydataset_latam[['pf_identity','hf_score']].std(),
           ydataset[['pf_identity','hf_score']].std()],axis=1).rename(columns={0:'D.E LatAm',
                                                                               1:'D.E Mundo'})
```

Calcular la moda tiene utilidad limitada. En principio puede ser útil para identificar una tendencia regional, pero en este caso pierde sentido práctico ya que los índices dependen del promedio de otras variables, comportándose como un dato continuo.

### 2- Realizar una prueba de Kolmogorov-Smirnof para comprobar analíticamente si estas variables responden la distribución propuesta en el ejercicio anterior. Hint: podés usar https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.kstest.html, pero hay que tener en cuenta que si la distribución es "norm", entonces va a comparar los datos con una distribución normal con media 0 y desviación estándar 1. Se puede utilizar la distribución sobre todos los datos o sólo sobre Latinoamérica.

Tomaremos la muestra de Latinoamérica.

```python
#H0: Los datos analizados siguen una distribución M.
#H1: Los datos analizados no siguen una distribución M.
#z=(valor observado - media ) / varianza

#primero ajusto mean = 0 y std = 1 de cada variable para mi hipótesis H0.
hip_mean_pf, hip_std_pf = st.norm.fit(ydataset_latam.pf_identity)
hip_mean_hf, hip_std_hf = st.norm.fit(ydataset_latam.hf_score)

#aplico la fórmula
zlatam_pf = (ydataset_latam.pf_identity - hip_mean_pf)/hip_std_pf
zlatam_hf = (ydataset_latam.hf_score - hip_mean_hf)/hip_std_hf
```

```python
st.kstest(zlatam_pf, 'norm')
```

```python
st.kstest(zlatam_hf, 'norm')
```

Observamos que tanto para el hf_score como para el pf_identity el p-value es distinto de 0, por lo tanto no podemos rechazar la hipótesis 0.

### 3- Realizar un gráfico QQ de las mismas distribuciones. Se puede utilizar ambas distribuciones sobre todos los datos o sólo sobre Latinoamérica, pero no cruzadas.

```python
plt.figure(figsize=(12,8))
#Nueva distribución normal con los parametros anteriores
nd_latam_pf = st.norm(hip_mean_pf,hip_std_pf)
#Generamos muestra
latamsample = nd_latam_pf.rvs(len(ydataset_latam))##len = 26
#Graficamos
seaborn.distplot(latamsample, label='Distribución normal')
seaborn.distplot(ydataset_latam['pf_identity'], label='pf_identity')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
seaborn.despine()
```
